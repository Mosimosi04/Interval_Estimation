```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def plot_t_distribution(bias, alpha, df, se):
    # 計算t分配區間
    t_value = stats.t.ppf(1 - alpha / 2, df)
    ci_low = bias - t_value * se
    ci_high = bias + t_value * se

    # 計算t分配曲線
    x = np.linspace(bias - 4 * se, bias + 4 * se, 100)
    y = stats.t.pdf(x, df, loc=bias, scale=se)

    # 繪製t分配曲線
    plt.plot(x, y)
    
    # 繪製偏倚為0的垂直線
    plt.axvline(x=0, color='black', linestyle=':')
    
    # 繪製中心值的垂直線
    plt.axvline(x=bias, color='g')
    plt.text(bias, 0.1, f'bias: { bias:.4f}', color='g')

    # 繪製信賴區間的垂直線
    plt.axvline(x=ci_low, color='b', linestyle='--')
    plt.axvline(x=ci_high, color='b', linestyle='--')
    plt.text(ci_low, 7, f'CI Low: {ci_low:.4f}', color='black')
    plt.text(ci_high, 7, f'CI High: {ci_high:.4f}', color='black')
    
    #合格管制圖法使用的，沒用就先備註掉
    #plt.axvline(x=0.0067, color='black', linestyle='--')
    #plt.text(0.0067, 4, f'xbar: {0.0067:.4f}', color='black')
    
    #故意NG管制圖法使用的，沒用就先備註掉
    plt.axvline(x=0.13, color='black', linestyle='--')
    plt.text(0.13, 4, f'xbar: {0.13:.4f}', color='black')
    
    # 繪製信賴區間的水平線並加上箭頭
    plt.annotate('', xy=(ci_low, 2), xytext=(ci_high, 2), 
                 arrowprops={'arrowstyle': '<->', 'color': 'b', 'linewidth': 0.5})

    # 設定圖形標籤
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.title('t-Distribution')

    # 顯示圖形
    plt.show()
```

```python
# 設定引數值
bias = 0.026
alpha = 0.05
df = 36.5
se = 0.0459

# 繪製t分配圖形
plot_t_distribution(bias, alpha, df, se)
```

```python
# 故意弄顯著的案例，把bias調整為0.1
bias = 0.1
alpha = 0.05
df = 36.5
se = 0.0459

# 繪製t分配圖形
plot_t_distribution(bias, alpha, df, se)
```

```python
# 管制圖法，直接偏倚=0，先算出上下限
bias = 0
alpha = 0.05
df = 36.5
se = 0.0459

# 繪製t分配圖形
plot_t_distribution(bias, alpha, df, se)
```

```python
"""
Python中的scipy.stats.t函数可以计算自由度为非整数的t分布。在scipy.stats.t中，自由度参数df可以是任意实数。
当自由度为整数时，scipy.stats.t返回的结果将与Excel的T.INV.2T函数相同
"""
import scipy.stats as stats
alpha = 0.05
df = 36.5
crit_val = stats.t.ppf(1 - alpha/2, df)
print (crit_val)
```

```python
point_estimate = 0.026
se = 0.0459
lower_bound = point_estimate - crit_val * se
upper_bound = point_estimate + crit_val * se
print(f"The {1-alpha:.0%} confidence interval is [{lower_bound:.4f}, {upper_bound:.4f}].")
```

```python
# 設定引數值，bias 獨立樣本法案例
bias = 0.0067
alpha = 0.05
df = 14
se = 0.2120/(15**0.5)
# 繪製t分配圖形
plot_t_distribution(bias, alpha, df, se)
```
